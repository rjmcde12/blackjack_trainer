```python
import os
```

```python
old_files = os.listdir()
old_files = sorted(old_files)
```

```python
new_files = []
for file in old_files:
    new_file = file.replace('diamonds', 'd')
    new_file = new_file.replace('hearts', 'h')
    new_file = new_file.replace('clubs', 'c')
    new_file = new_file.replace('spades', 's')
    new_file = new_file.replace('king', 'K')
    new_file = new_file.replace('queen', 'Q')
    new_file = new_file.replace('jack', 'J')
    new_file = new_file.replace('ace', 'A')
    new_file = new_file.replace('_of_', '')
    new_files.append(new_file)
    
new_files = sorted(new_files)
del new_files[0:2]
del new_files[-1]
```
